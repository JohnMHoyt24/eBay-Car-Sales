What caught my attention are the standard deviations for the respective columns. I am curious to see why the standard deviation for odometer_km is lower than the standard deviation for price, and what that says about the data.

According to the _National Library of Medicine_ (https://www.nlm.nih.gov/nichsr/stats_tutorial/section2/mod8_sd.html), "Low standard deviation means data are around the mean, and high standard deviation indicates data are more spread out." This can explain why odometer_km has very few unique values, in comparison to price. The standard deviation for odometer_km is low and most of its data is centered around the mean. 

With price, the data is spread out, and not clustered around the mean, that's because price has a much higher standard deviation than odometer_km. But, because price's data is more spread out, could there be the possibility of outliers in the data? Normally, outliers tend to be spread out and isolated from the other data points.

```python
new_autos["odometer_km"].value_counts(ascending=True)
```

```python
new_autos["price"].value_counts().head()
```

value_counts() shows that the data for odometer_km is mostly centered around the mean. Very unlikely there's any outliers in that column. value_counts() for price, on the other hand, shows values varying drastically. This is a sign that there could be outliers in the price column.

```python
top_brand_by_year = {}

auto_brands = ry_autos['brand'].unique()
```

```python
for a in auto_brands:
    selected_rows = ry_autos[ry_autos['brand'] == a]
    sorted_rows = selected_rows.sort_values('date_crawled', ascending=False)
    brand = sorted_rows.iloc[7]['registration_year']
    top_brand_by_year[a]= brand
```

This for loop will aggregate data pertaining to the year and its top automobile brand. The rows are sorted in accordance with the date_crawled column.
